```python
#genderclean_df["AgeBin"]='0-9'

def add_age_bin(row):
    if row['AgeStart']>=0 and row['AgeStart']<10:
        return "00-10"
    elif row['AgeStart']>=10 and row['AgeStart']<20:
        return "10-19"
    elif row['AgeStart']>=20 and row['AgeStart']<30:
        return "20-29"
    elif row['AgeStart']>=30 and row['AgeStart']<40:
        return "30-39"
    elif row['AgeStart']>=40 and row['AgeStart']<50:
        return "40-49"
    elif row['AgeStart']>=50 and row['AgeStart']<60:
        return "50-59"
    elif row['AgeStart']>=60 and row['AgeStart']<70:
        return "60-69"
    elif row['AgeStart']>=70 and row['AgeStart']<80:
        return "70-79"
    else:
        return "80+"
genderclean_df["AgeBin"] = genderclean_df.apply(lambda row:add_age_bin(row), axis=1)
genderclean_df
```
